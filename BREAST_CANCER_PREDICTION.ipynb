```python
df=df.dropna(axis=1)
# dropna() dunction is used to remove missing values
#      axis=0  --> Drop rows which contain missing values.
#      axis=1  --> Drop columns which contain missing value.
```

```python
df.shape
# WE CHECK THE SIZE OF THE DATASET AFTER REMOVING THE NULL COLUMNS
```

```python
df.describe()

#The describe() method is used for calculating some statistical data like 
#percentile, mean and std of the numerical values of the Series or DataFrame. 
#It analyzes both numeric and object series and also the DataFrame column sets of mixed data types
```

```python
# GET THE COUNT OF MALIGNANT(M) AND BENIGN(B) CCELLS

df['diagnosis'].value_counts()
```

```python
sns.countplot(df['diagnosis'],label="count")
```

```python
# LABEL ENCODING ( CONVERT THE VALUE OF M AND B INTO 1 AND 0)
from sklearn.preprocessing import LabelEncoder
labelencoder_Y = LabelEncoder()
df.iloc[:,1]=labelencoder_Y.fit_transform(df.iloc[:,1].values)

# iloc() function enables us to select a particular cell of the dataset
```

```python
df.head()
```

```python
# GET THE CORRELATION


# If two variables are closely correlated, then we can predict one variable from the other

df.iloc[:,1:32].corr()

```

```python
# VISUALIZE THE CORRELATION

plt.figure(figsize=(10,10))
sns.heatmap(df.iloc[:,1:10].corr(),annot=True,fmt=".0%")
```

```python
# SPLITTING THE DATASET INTO DEPENDENT(X) AND INDEPENDENT(Y)  DATASETS
X=df.iloc[:,2:31].values
Y=df.iloc[:,1].values

# .values    Return a Numpy representation of the DataFrame.
```

```python
# Splitting the data into training and test dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

# MODELS / ALGORITHMS

## Random Forest 



*    n_estimators : This is the number of trees you want to build before taking the maximum voting or averages of predictions. Higher number of trees give you better performance but makes your code slower.




```python
from sklearn.ensemble import RandomForestClassifier
RFC=RandomForestClassifier(n_estimators=200)
RFC.fit(X_train,Y_train)
```



---
TESTING THE DATASET FOR TRAIN , CHECK IF WE GET 100 % ACCURACY



```python
RFC_train= RFC.predict(X_train)
from sklearn import metrics
print("Accuracy Score = ",format(metrics.accuracy_score(Y_train,RFC_train)))
```



--- 
TESTING THE TEST DATASET 



```python
from sklearn import metrics 
predictions= RFC.predict(X_test)
print("Accuracy Score = ",format(metrics.accuracy_score(Y_test,predictions)))
```



---
PRINTING THE CONFUSION MATRIX



```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(Y_test,predictions))
print("------------------------------------------------")
print(classification_report(Y_test,predictions))
                       
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(random_state=0,max_iter=10000)
LR.fit(X_train,Y_train)
```

```python
LR_train= RFC.predict(X_train)
from sklearn import metrics
print("Accuracy Score = ",format(metrics.accuracy_score(Y_train,LR_train)))
```

```python
from sklearn import metrics 
predictions= LR.predict(X_test)
print("Accuracy Score = ",format(metrics.accuracy_score(Y_test,predictions)))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(Y_test,predictions))
print("------------------------------------------------")
print(classification_report(Y_test,predictions))
                       
```

## DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
DTC= DecisionTreeClassifier(random_state=0,criterion="entropy")
DTC.fit(X_train,Y_train)
```

```python
LR_train= DTC.predict(X_train)
from sklearn import metrics
print("Accuracy Score = ",format(metrics.accuracy_score(Y_train,LR_train)))
```

```python
from sklearn import metrics 
predictions= DTC.predict(X_test)
print("Accuracy Score = ",format(metrics.accuracy_score(Y_test,predictions)))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(Y_test,predictions))
print("------------------------------------------------")
print(classification_report(Y_test,predictions))
                       
```
